# LogSense — Data Exploration
Exploratory analysis of the cybersecurity intrusion dataset.
This notebook covers loading, inspecting, and understanding the data before modelling.

```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/cybersecurity_intrusion_data.csv')
print("Dataset loaded successfully")
print(f"Shape: {df.shape}")
df.head()
```

```python
# Check column names, data types and missing values
print("=== Column Info ===")
print(df.dtypes)
print("\n=== Missing Values ===")
print(df.isnull().sum())
```

```python
# Check how many attacks vs normal sessions we have
print("=== Target Distribution ===")
print(df['attack_detected'].value_counts())
print("\n=== As Percentage ===")
print(df['attack_detected'].value_counts(normalize=True).round(2) * 100)
```

```python
# Look at the text columns to see what unique values they contain
text_columns = ['protocol_type', 'encryption_used', 'browser_type']

for col in text_columns:
    print(f"=== {col} ===")
    print(df[col].value_counts())
    print()
```

## Data Summary
Dataset is clean and ready for preprocessing with one exception:
- `encryption_used` has 1,966 missing values — will be filled with "None"
- All text columns have small, manageable category counts
- Target is well balanced (55% benign / 45% attack)
